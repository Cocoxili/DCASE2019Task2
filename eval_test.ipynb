```python
predicted.columns = ['fname', 'pred']

actual = actual[['fname', 'label']]

data_all = pd.merge(actual, predicted, how = 'left', on = 'fname')

data_all['values'] = data_all['label'].values + '+' + data_all['pred'].values

data_all['map@3'] = data_all['values'].\
apply(lambda x:metrics.apk([x.split('+')[0]], list(x.split('+')[1].split(" ")) ,3)).values

data_all[['fname', 'map@3']]

data_all
```
